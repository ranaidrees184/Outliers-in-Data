```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('weight-height.csv')
```

```python
df.sample(3)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.boxplot(df['Height'])
plt.subplot(1,2,2)
sns.distplot(df['Height'],hist=False)
```

```python
upperlimit=df['Height'].quantile(0.99)
upperlimit
```

```python
lowerlimit=df['Height'].quantile(0.01)
lowerlimit
```

# Remove Values (Trimming)

```python
remdf=df[(df['Height']<=upperlimit) & (df['Height']>=lowerlimit)]
```

```python
sns.boxplot(remdf['Height'])
```

# Winserization or Capping

```python
df['Height']=np.where(df['Height']>upperlimit,upperlimit,np.where(df['Height']<lowerlimit,lowerlimit,df['Height']))
df
```
